```python
import os 
import pandas as pd
Label_data = pd.read_excel("Acoustic_dataset.xlsx")
Demographic_data = pd.read_excel("BERT_RF.py/EMA_Dataset.xlsx")
```

```python
# Extract specific columns: ID, Label, Sex, Age from EMA dataset
selected_columns = Demographic_data[['ID', 'Label', 'Sex', 'Age']]
print("Extracted columns from EMA dataset:")
print(f"Shape: {selected_columns.shape}")
selected_columns
```

```python
data = Label_data
```

```python
# Merge the selected demographic data with the acoustic data (Label_data)
complete_data = pd.merge(selected_columns, Label_data, on='ID', how='inner')

print("Merged complete dataset:")
print(f"Shape: {complete_data.shape}")
print(f"Columns: {complete_data.columns.tolist()}")
print("\nFirst few rows:")
complete_data.head()
```

```python
users = list(set(data["ID"]))

# 打印一下数据基本情况
print(f"总用户数: {len(users)}，总样本数: {len(data)}")

```

```python
# Save the complete_data to a CSV file
complete_data.to_csv('complete_data.csv', index=False)
print(f"Data saved to complete_data.csv")
print(f"File contains {complete_data.shape[0]} rows and {complete_data.shape[1]} columns")
```
